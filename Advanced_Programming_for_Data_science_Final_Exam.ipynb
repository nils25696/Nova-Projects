```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.cluster import SpectralClustering
import warnings
warnings.filterwarnings("ignore")
```

```python
# Exploration: 

# combination of directives and characters needed to read tsla: '%Y+%m+%d'
tsla = pd.read_csv('TSLA.csv', parse_dates=['Date'], date_parser=lambda x: pd.datetime.strptime(x, '%Y+%m+%d'))

# combination of directives and characters needed to read tsla: '%y--%m--%d'
albe = pd.read_csv('ALBE.csv', parse_dates=['Date'], date_parser=lambda x: pd.datetime.strptime(x, '%y--%m--%d'))
```

```python
class Stockgo():
    """
    Speeds up analyses.

    Attributes:
        filename (str): The name of the file to load.
        directives (str): The format of the date and time information in the file.
        label (str): A label for the stock data.
        trimmed (bool): A flag to indicate whether the data should be trimmed.
        
    Methods:
        report(self, column='High')
        trim(self, start_date='2011-01-01', end_date='2017-01-01')
    """

    def __init__(self, filename, directives='%Y-%m-%d %H:%M:%S', label='stock', trimmed=False):
        # initialize the attributes with the provided values
        self.filename = filename
        self.directives = directives
        self.label = label
        self.trimmed = trimmed

        # try to read the file into a DataFrame and raise an error if the file is not found
        try:
            self.df = pd.read_csv(self.filename)
        except FileNotFoundError:
            raise FileNotFoundError(f"File '{self.filename}' not found")

        # convert the temporal_info column to datetime format using the provided directives
        self.df['Date'] = pd.to_datetime(self.df['Date'], format=self.directives)

        # set the temporal_info column as the index of the DataFrame
        self.df.set_index('Date', inplace=True)

    def report(self, column='High'):
        """
        Print statistical properties and plot the chosen column of the stock data.
        """
        # print the statistical properties of the DataFrame
        print(self.df.describe())

        # plot the chosen column with the appropriate label and axis
        ax = self.df[column].plot()
        ax.set_ylabel(column)
        ax.legend([self.label])

    def trim(self, start_date='2011-01-01', end_date='2017-01-01'):
        """
        Filter out unwanted data from the stock data between the start and end dates.
        """
        # validate the start_date and end_date arguments
        if not isinstance(start_date, str) or not isinstance(end_date, str):
            raise TypeError("Both temporal arguments must be strings")
        if len(start_date) != 10 or len(end_date) != 10:
            raise Exception("Both arguments must have length 10")

        # convert the start_date and end_date to datetime objects
        try:
            start_date = pd.to_datetime(start_date, format='%Y-%m-%d')
        except ValueError:
            raise Exception("Can't convert start_date to temporal format")
        try:
            end_date = pd.to_datetime(end_date, format='%Y-%m-%d')
        except ValueError:
            raise Exception("Can't convert end_date to temporal format")

        # filter the DataFrame by the start_date and end_date
        self.df = self.df[(self.df.index >= start_date) & (self.df.index <= end_date)]

        # set the trimmed attribute to True
        self.trimmed = True
```

```python
# create an instance of Stockgo for the "ALBE.csv" file
ca = Stockgo('ALBE.csv', directives='%y--%m--%d', label='ALBE')
```

```python
# create an instance of Stockgo for the "TSLA.csv" file
ct = Stockgo('TSLA.csv', directives='%Y+%m+%d', label='TSLA')
```

```python
# trim the data with default arguments
ca.trim()
ct.trim()
```

```python
# use the report method to compare the 'Open' mean and 25th percentile for each stock
print("ALBE:")
ca.report(column='Open')
```

```python
print("\nTSLA:")
ct.report(column='Open')
```

```python
# compare the mean and 25th percentile values for each stock
if ca.df['Open'].mean() > ct.df['Open'].mean():
    print("ALBE has a higher mean 'Open'")
else:
    print("TSLA has a higher mean 'Open'")

if np.percentile(ca.df['Open'], 25) > np.percentile(ct.df['Open'], 25):
    print("ALBE has a higher 25th percentile 'Open'")
else:
    print("TSLA has a higher 25th percentile 'Open'")
```

```python
# make a scatter plot of the High values of ca(x-axis) and ct(y-axis)
plt.scatter(ca.df['High'], ct.df['High'])
plt.xlabel('ALBE High')
plt.ylabel('TSLA High')
plt.show()
```

**Conclusion:** There seem to be three different clusters here.

```python
# use Spectral Clustering to cluster the data
X = np.vstack((ca.df['High'], ct.df['High'])).T
scl = SpectralClustering(affinity='nearest_neighbors', n_clusters=3)
scl.fit(X)
```

```python
# make a scatter plot of the High values of ca(x-axis) and ct(y-axis) with cluster labels
plt.scatter(ca.df['High'], ct.df['High'], c=scl.labels_)
plt.xlabel('ALBE High')
plt.ylabel('TSLA High')
plt.show()
```

```python
# make a scatter plot of the High values of ca and ct over time with cluster labels
plt.scatter(ca.df.index, ca.df['High'], c=scl.labels_)
plt.scatter(ct.df.index, ct.df['High'], c=scl.labels_)
plt.xlabel('Date')
plt.ylabel('High')
plt.show()
```
